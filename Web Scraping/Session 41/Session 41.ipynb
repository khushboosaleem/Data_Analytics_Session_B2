```python
columns = table.find_all('th')
```

```python
columns
```

```python
for i in columns:
    print(i.text.strip())
```

```python
columns_title = [i.text.strip() for i in columns]
```

```python
columns_title
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(columns = columns_title)
```

```python
df
```

```python
columns_data = table.find_all('tr')
```

```python
columns_data
```

```python
columns_data[1:]
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    print(row_data)
```

```python
for row in columns_data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    print(ind_row_data)
    length = len(df)
    df.loc[length] = ind_row_data
    
```
