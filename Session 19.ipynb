```python
lessthan5 = []     
fiveto10 = []
tento15 = []
fifteento20 = []
twentyto25 = []
moretha25 = []


for i in survey:
    if i[-1] == "< 5 Years":
        lessthan5.append(i[-2])
    elif i[-1] == "5-10 Years":
        fiveto10.append(i[-2])
    elif i[-1] == "10-15 Years":
        tento15.append(i[-2])
    elif i[-1] == "15-20 Years":
        fifteento20.append(i[-2])
    elif i[-1] == "20-25 Years":
        twentyto25.append(i[-2])
    else:
        moretha25.append(i[-2])
```
