```python
# import important libraries
import pandas as pd
import sqlite3 as sql
```

```python
# make connection
conn = sql.connect("chinook.db")

def sql_query(q):
    return pd.read_sql_query(q, conn)
    
```

```python
 sql_query("select * from sqlite_master")
```

```python
# check the table names
sql_query("select * from sqlite_master where type = 'table'")
```

```python
# invoice_id, track_id, track_name, media_type name, quantity, unit_price  where invoice_id = 1

q = """select il.invoice_id, t.track_id, t.name track_name, m.name media_name, il.quantity, il.unit_price
        from invoice_line il
        inner join track t
        on 
        il.track_id = t.track_id
        inner join media_type m
        on 
        t.media_type_id = m.media_type_id
        where il.invoice_id == 1

"""

sql_query(q)
```
